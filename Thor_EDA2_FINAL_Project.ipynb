```python
#FEATURE ENGINEERED ANOTHER VARIABLE ACTUAL LEAD TIME
PO_EDA1.shape
PO_EDA1['ACTUAL LEAD TIME'] = PO_EDA1['RECEIVED'] - PO_EDA1['DOCK DATE']
PO_EDA1.info()

```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':100})

# plot
plt.boxplot(PO_EDA1['QUANTITY ORDERED'], boxprops=dict(color='red'), showfliers=False)
plt.title('QUANTITY ORDERED')
PO_EDA1.describe()
#boxplot = PO_EDA1.boxplot(column=['QUANTITY ORDERED','QUANTITY RECEIVED','ACTUAL LEAD TIME'])
```

```python
#FEATURE ENGINEERED DIFFERENCE BETWEEN QUANTITY ORDERED AND QUANTITY RECEIVED
PO_EDA.info()
PO_EDA1['QUANTITY RECEIVED DELTA'] = PO_EDA1['QUANTITY ORDERED'] - PO_EDA1['QUANTITY RECEIVED']
PO_EDA1.info()
PO_EDA1.head()
```

```python
##REMOVING OUTLIERS FROM QUANTITY ORDERED; WITH QUANTITY THAT SHOULD BE GREATHER THAN 1
q_low = PO_EDA1["QUANTITY ORDERED"].quantile(0.01)
print(q_low)
q_hi  = PO_EDA1["QUANTITY ORDERED"].quantile(0.99)
print(q_hi)

df_filtered = PO_EDA1[(PO_EDA1["QUANTITY ORDERED"] < q_hi) & (PO_EDA1["QUANTITY ORDERED"] >= 1)]
df_filtered.head()
df_filtered.info()
df_filtered.tail()
#REMOVED 872 OUTLIERS THAT HAS QUANTITY ORDERD MORE THAN TOP 1% (450)
```

```python
##REMOVING OUTLIERS FROM ACTUAL LEAD TIME
q_low_days = PO_EDA1["ACTUAL LEAD TIME"].dt.days.quantile(0.01)
print(q_low)
q_hi_days  = PO_EDA1["ACTUAL LEAD TIME"].dt.days.quantile(0.99)
print(q_hi)

df_filtered1 = df_filtered[(df_filtered["ACTUAL LEAD TIME"].dt.days < q_hi_days) & (df_filtered["ACTUAL LEAD TIME"].dt.days > q_low_days)]
df_filtered1.head()
df_filtered1.info()
df_filtered1.tail()
##REMOVED 1744 ROWS THAT HAS ACTUAL LEAD TIME LESS THAN 1 AND top 1% actual lead time days
```

```python
df_filtered1.info()
df_filtered1.tail()
df_filtered1.describe()
```

```python
##REMOVING QUANTITY RECEIVED DELTA OUTLIERS;
q_low_receive = df_filtered1["QUANTITY RECEIVED DELTA"].quantile(0.001)
print(q_low_receive)
q_hi_receive  = df_filtered1["QUANTITY RECEIVED DELTA"].quantile(0.999)
print(q_hi_receive)

df_filtered2 = df_filtered1[(df_filtered1["QUANTITY RECEIVED DELTA"] < q_hi_receive) & (df_filtered1["QUANTITY RECEIVED DELTA"] > q_low_receive)]
df_filtered2.head()
df_filtered2.info()
df_filtered2.tail()
df_filtered2.shape
df_filtered2.describe()
##REMOVED 177 OUTLIER ROWS OF QUANTITY RECEIVED DELTA 
```
